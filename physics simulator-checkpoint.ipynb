```python
import matplotlib.pyplot as plt
import numpy as np

# Enable inline plotting in JupyterLab
%matplotlib inline  

# Simple Car Motion Simulator with Visualization
def car_motion():
    # User inputs
    mass = float(input("Enter car mass (kg): "))
    power = float(input("Enter engine power (W): "))
    time = float(input("Enter acceleration time (s): "))

    # Initial conditions
    initial_velocity = 0  

    # Approximate force using Power / velocity (assuming avg velocity = final_velocity / 2)
    avg_velocity = power / mass  
    force = power / avg_velocity  

    # Calculate acceleration (F = ma)
    acceleration = force / mass  

    # Calculate velocity over time
    time_values = np.linspace(0, time, 100)  # Creates 100 time points
    velocity_values = initial_velocity + (acceleration * time_values)

    # Final velocity
    final_velocity = velocity_values[-1]

    # Calculate stopping distance (vÂ² / 2a)
    stopping_distance = (final_velocity ** 2) / (2 * acceleration)

    # Plot velocity vs. time
    plt.figure(figsize=(8, 5))
    plt.plot(time_values, velocity_values, label="Velocity (m/s)", color="b")
    plt.axhline(y=final_velocity, color='r', linestyle="--", label=f"Final Velocity: {final_velocity:.2f} m/s")
    plt.axvline(x=time, color='g', linestyle="--", label=f"Acceleration Time: {time} s")

    # Labels and title
    plt.xlabel("Time (s)")
    plt.ylabel("Velocity (m/s)")
    plt.title("Car Velocity Over Time")
    plt.legend()
    plt.grid(True)

    # Show the plot in Jupyter
    plt.show()

    # Print results in Jupyter
    from IPython.display import display, Markdown
    display(Markdown(f"### ðŸš— Simulation Results"))
    display(Markdown(f"**Final Velocity:** {final_velocity:.2f} m/s"))
    display(Markdown(f"**Acceleration:** {acceleration:.2f} m/sÂ²"))
    display(Markdown(f"**Stopping Distance:** {stopping_distance:.2f} meters"))

# Run the simulator
car_motion()
```
